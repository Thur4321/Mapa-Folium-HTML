```python
import folium
import pandas as pd
from folium import plugins
```

```python
m = folium.Map(location=[-23.547,-46.635], tiles='Stamen Toner', zoom_start=7)
df = pd.read_csv("data.csv", sep = ';', usecols = ['municipio', 'nomeEstacao', 'latitude', 'datahora'])
df1 = df.replace({',':'.'}, regex = True)
df1['datahora'] = df1['datahora'].astype(float, errors = 'raise')
coordenadas = []
df2 = df1.groupby(['municipio', 'nomeEstacao', 'latitude']).agg({'datahora':'sum'}).reset_index()
group = folium.FeatureGroup('Locais')
```

```python
for lat, lon, pes, name in zip(df2.nomeEstacao, df2.latitude, df2.datahora, df2.municipio):
    coordenadas.append([lon, lat, pes])
    group.add_child(folium.Marker(location = [lon, lat], popup = f'Estação: '+name+'\nQuantidade (em mm): '+str(round(pes, 2))))
```

```python
m.add_child(plugins.HeatMap(coordenadas))
m.add_child(group)
```

```python
m.save('mapa.html')
```
